### Renaming column names
The original column names have spaces (and are long-ish), which may make the dataframe a little harder to work with. Let's rename them.

To rename columns names, we can assign a list of new column names to the dataframe's `columns` attribute. 

If we want to rename only one or a few of the columns, it's handy to use the `.rename` method. 

```python
list_rename = ['year', 'month', 'day', 'timeofday', 'dayofweek', 'city', 'intersection', 'hitandrun', 
                  'lighting', 'weather', 'speedlimit', 'worstinjury', 'partytype', 'age', 'gender']

crash.rename(columns={'year':'Year', 'month':'Month', 'day':'Day'})
```

```python
crash.columns = list_rename
crash.head()
```

```python
crash.head(3)
```

### Replacing space by underoscore (_)

```python
crash.columns = crash.columns.str.replace(' ','_')
crash.columns
```

### Checking missing values
In many real-world datasets you'll find missing values. These can interfere with data analysis processes, so we check for these first.

```python
crash.isnull().sum()
```

```python
crash.isnull().any()
```

```python
import numpy as np

crash.replace({'DOB invalid': np.nan,'Weather_Conditions_(2016+)	': np.nan,'Uncoded & errors':np.nan}, inplace = True)
```

```python
crash.head()
```

```python
crash.isnull().sum()
```

### Subsetting columns
To select a single column we can use something like either `crash.city` or `crash['city]`. 

To select multiple columns we can use something like `crash[['city', 'lighting', 'speedlimit']]`

```python
crash['city'][:5]
```

```python
crash.age
```

```python
crash[['city', 'lighting', 'speedlimit']][:5]
```

### Value counts
Pandas `value_counts()` method gives us the number of times each value shows up in the given column (often called a pandas Series).

```python
crash.gender.value_counts()
```

### Number of unique values
Pandas `nunique()` method gives us the number of unique values in a column

```python
crash['month'].nunique()

crash['day'].nunique()
```

```python
# unique will show us the unique od each with count
crash['month'].unique()
```

### Filtering a dataframe

```python
#To find all data point for the year 2018
crash[crash.year == 2018]
```

```python
# crash year beteween 2010,2013
crash[crash.year.between(2010,2013)]
```

```python
# let find crash for day one and month january
crash[(crash.day== 1)&(crash.month == 'January')]
```

### Creating a new column

```python
# To create a column named as NULL for demo
crash['NULL'] = crash['year']*2
crash.head()
```

### Deleting a column
Pandas `.drop` method allows us to delete a list of columns by name.

```python
# To drop the NULL column that we have create above
crash.drop('NULL',axis =1,  inplace =True)
crash[:3]
```

### Accessing a row
Pandas `.iloc` and `.loc` method allows us to access specified rows by position.

```python
crash.iloc[2:6]
```

```python
# To show the first 5 rows and first 5 attributes(columns)
# we can see here that the iloc work with index label
crash.iloc[:5,:5]
```

```python
# loc work with the label itself 
crash.loc[:4,['year','month',	'day',	'timeofday',	'dayofweek']]
```

```python
crash.iloc[:20,:10].describe()
```
